## if文

if文は、条件分岐を実現するための基本構文のひとつです。  
条件が成立した場合にコードを実行し、不成立の場合には別のコードを実行します。


*2分岐*
```Python
if 条件式:
  条件式が成立したときの処理
else:
  条件式が不成立のときの処理  
```

*３分岐以上*

```Python
if 条件式1:
  条件式1が成立したときの処理
elif 条件式２:
  条件式2が成立したときの処理
…
elif 条件式n:
  条件式nが成立したときの処理
else:
  全ての条件式が不成立のときの処理
```

### 実装例

````python
# テストの点数の設定
score = 65

# 条件分岐の構文
if score > 40:
    print('合格です！')
else:
    print('あなたは赤点です。')
````

````python
# テストの点数によりクラス分けをします
#点数を入力
score = int(input('あなたの点数を教えてください：'))

# 条件分岐の構文
if score >= 80:
    print('あなたは A クラスです')

elif score < 80 and score >= 50:
    print('あなたは B　クラスです。')

else:
    print('あなたは C　クラスです。')
````

## for文

for文は繰り返し処理を行う基本構文の一つです。  
同じ処理を何回も実行できます。


*リスト内の全要素を参照する*
```Python
for 変数 in リスト;
  繰り返し処理
```
*指定回数繰り返す*
```Python
for 変数 in range(n):
  繰り返し処理  
  （n回処理を実行）
```
*（応用）ネスト構造にできる*  
※使用頻度：高
```Python
for 変数１ in range(n):
    for 変数２ in range(m):
（n×m回処理を実行）
```

### 実装例

````python
# リストを定義する
names = ['太郎', '花子', '一郎']

# for 文
for name in names:
    print(name +'さん')
````

````python
#1~10までの和を求める
sum = 0
for i in range(11):
  sum += i
print(f'合計値は{sum}です')


#別解１
sum2 = 0
for i in range(1,11):
  sum2 += i
print(f'合計値は{sum2}です')
````

````python
#for文のネストを使って左した直角の直角三角形の作成
# 直角三角形の高さを設定
height = 5

for i in range(1, height + 1):
    # アスタリスクを出力するためのネストされたfor文
    for j in range(i):
      #改行しないようにend=''をいれている
        print('*', end='')
    # 改行を出力
    print()
````

###問１（中級）

BMIを求めよ  
以下の式と表を参考にしてBMIの値と、肥満度判定の結果を出力せよ。  
※体重，身長はコード内で定義して良いものとする。  

$
BMI =　[体重(kg)] ÷ [身長m)]^2
$


 BMIによる肥満度の対応表

| BMI範囲       | 肥満度分類  |
|---------------|--------------|
| 18.5未満      | 低体重（痩せ型） |
| 18.5 - 24.9   | 普通体重    |
| 25 - 29.9     | 肥満（1度） |
| 30 - 34.9     | 肥満（2度） |
| 35 - 39.9     | 肥満（3度） |
| 40以上        | 肥満（4度） |

※BMIは体重と身長から算出される肥満度を表す体格指数です

参考資料  
[BMIとは](https://www.e-healthnet.mhlw.go.jp/information/dictionary/metabolic/ym-002.html)

### 解答例

````python
# 体重（kg）と身長（m）をコード内で定義
weight = 70  # 例: 70kg
height = 1.75  # 例: 1.75m

# BMIの計算
bmi = weight / (height ** 2)

# 肥満度判定
if bmi < 18.5:
    category = "低体重（痩せ型）"
elif 18.5 <= bmi < 25:
    category = "普通体重"
elif 25 <= bmi < 30:
    category = "肥満（1度）"
elif 30 <= bmi < 35:
    category = "肥満（2度）"
elif 35 <= bmi < 40:
    category = "肥満（3度）"
else:
    category = "肥満（4度）"

# 結果の出力
print(f"BMIの値: {bmi:.2f}")
print(f"肥満度判定: {category}")
````

### 問２
閏年を判定するプログラムを作成してください。

閏年は以下で判定できるものとする。
西暦はキーボードから入力を受け付けるものとする。

*   西暦が4で割り切れる年は閏年
*   西暦が100で割り切れる年は閏年ではない
*   西暦が400で割り切れる年は閏年



### 解答例

````python
# ユーザーから西暦年を入力として受け取る
year = int(input("西暦年を入力してください: "))
# 閏年判定
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(f"{year}年は閏年です")
        else:
            print(f"{year}年は閏年ではありません")
    else:
        print(f"{year}年は閏年です")
else:
    print(f"{year}年は閏年ではありません")
````

## 上級問題


モンテカルロ法を用いて円周率πの近似値を計算してください
モンテカルロ法は、大量のランダムなサンプル（乱数）を生成し、そのサンプルをもとに問題の解答を近似的に求める方法です。  

計算手順

1.   １辺の長さが１の正方形の中に、ランダムに点を打つ
2.   正方形の左下から距離が１以下の点の総数をmとする
3.   ランダムな点の総数をnとする
4.  正方形の面積は１なので、m/nが1/4円の面積の近似値である
5.  円の半径は1なので、m/nを4倍した値がπの近似値になる

ヒント：
ランダムな数字の作成例
```Python
import random
num = random.random()

```

参考

### 解答例

````python
import random

# ランダムに点を打つ総数を指定
n = 1000000  # 任意の大きな数
m = 0

for i in range(n):
    # ランダムな点を生成
    x = random.random()
    y = random.random()

    # 点が円の内側にあるかどうかを判定
    if x**2 + y**2 <= 1:
        m += 1

# πの近似値を計算
pi_ = (m / n) * 4

# 結果を出力
print("円周率πの近似値:", pi_)
````
