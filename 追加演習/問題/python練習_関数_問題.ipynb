## 出力（戻り値）のある関数

```python
# 戻り値がない場合
def calc(a, b):
  print(a+b)
```

```python
result = calc(a=1, b=2)
```

```python
# result に格納されているのか・・？
result
```

```python
# 他の確認方法
type(result)
```

戻り値を用意した場合

```python
# 戻り値を用意する
def calc(a, b):
  return a+b
```

```python
result = calc(a=1, b=2 )
```

```python
result
```

```python
# 戻り値を複数設定
def calc(a, b):
  return a+b, a-b
```

```python
result = calc(a=1, b=2)
result
```

```python
# 戻り値を複数の変数で設定
wa, sa = calc(a=1, b=2)
wa, sa
```
