問題18:
以下の数列の中で、偶数の平均値を計算してください。

10
,
23
,
45
,
68
,
79
,
82
,
90
,
99
10,23,45,68,79,82,90,99

```python
numbers = [10, 23, 45, 68, 79, 82, 90, 99]
even_sum = 0
even_count = 0
for num in numbers:
    if num % 2 == 0:
        even_sum += num
        even_count += 1
average = even_sum / even_count
print(average)
```

問題19:
以下の数列の中で、最大値と最小値の差を計算してください。

4
,
78
,
23
,
56
,
89
,
12
,
45
,
67
4,78,23,56,89,12,45,67

```python
numbers = [4, 78, 23, 56, 89, 12, 45, 67]
max_val = numbers[0]
min_val = numbers[0]
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
difference = max_val - min_val
print(difference)
```

問題20:
1から50までの数字の中で、3の倍数でも5の倍数でもない数字の合計を計算してください。

```python
total = 0
for x in range(1, 51):
    if x % 3 != 0 and x % 5 != 0:
        total += x
print(total)
```
