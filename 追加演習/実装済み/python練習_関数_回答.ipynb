```python
# 戻り値を複数設定
def calc(a, b):
  return a+b, a-b
```

問題4:
与えられたリストの中で最も小さい数字を見つける関数 find_min を作成してください。

```python
def find_min(nums):
    smallest = nums[0]
    for num in nums:
        if num < smallest:
            smallest = num
    return smallest

numbers = [12, 5, 83, 34, 9, 65]
print(find_min(numbers))
```

問題11:

```python
def calculate_factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(calculate_factorial(5))
```

問題14:
与えられたリストの数字の平均を計算する関数 calculate_average を作成してください。

```python
def calculate_average(nums):
    total = 0
    for num in nums:
        total += num
    return total / len(nums)

numbers = [4, 8, 15, 16, 23, 42]
print(calculate_average(numbers))
```

問題15:
与えられた2つの数字の最小公倍数を返す関数 find_lcm を作成してください。

```python
def find_lcm(a, b):
    if a > b:
        greater = a
    else:
        greater = b

    while True:
        if greater % a == 0 and greater % b == 0:
            lcm = greater
            break
        greater += 1

    return lcm

print(find_lcm(4, 5))
```

問題16:
与えられた数字が素数かどうかを判定する関数 is_prime を作成してください。

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

print(is_prime(7))
```

問題18:
与えられた数字が完全数（自分自身を除く正の約数の和が自分自身と等しい数）であるかどうかを判定する関数 is_perfect を作成してください。

```python
def is_perfect(num):
    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i
    return total == num

print(is_perfect(28))
```

問題19:
与えられた2つの数字の最大公約数を返す関数 find_gcd を作成してください。

```python
def find_gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(find_gcd(12, 15))
```

問題20:
与えられた数字が回文数であるかどうかを判定する関数 is_palindrome_number を作成してください。

```python
def is_palindrome_number(num):
    return str(num) == str(num)[::-1]

print(is_palindrome_number(121))
```
